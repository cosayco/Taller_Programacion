## ¿Qué es Git y GitHub?
- **Git**: Sistema de control de versiones distribuido que permite gestionar el historial de cambios en proyectos.
- **GitHub**: Plataforma en la nube para alojar repositorios Git y colaborar en proyectos.

## Instalación de Git
Para instalar Git:
- **Windows**: Descargar desde [git-scm.com](https://git-scm.com).
- **Linux**: `sudo apt-get install git`
- **macOS**: `brew install git`

## Configuración inicial
Configura tu nombre y correo:
```bash
git config --global user.name "Tu Nombre"
git config --global user.email "tuemail@example.com"
```

## Comandos esenciales
- `git init`: Inicializa un repositorio.
- `git status`: Muestra el estado.
- `git add <archivo>`: Añade cambios al área de preparación.
- `git commit -m "mensaje"`: Guarda los cambios.
- `git push`: Envía cambios al remoto.
- `git pull`: Trae cambios del remoto.

## Conectar con GitHub
1. Crear un repositorio en GitHub.
2. Añadir el remoto:
```bash
git remote add origin https://github.com/usuario/repositorio.git
git branch -M main
git push -u origin main
```

## Flujo de trabajo típico
1. `git pull` para actualizar.
2. Editar archivos.
3. `git add` y `git commit`.
4. `git push` para subir cambios.
