1. üè∑Ô∏è**Label** ‚Äì Mostrar texto

```python
import tkinter as tk

ventana = tk.Tk()
ventana.title("Ejemplo de Label")

etiqueta = tk.Label(ventana, text="Bienvenido al sistema")
etiqueta.pack()

ventana.mainloop()
```

2. üìù **Entry** ‚Äì Caja de texto (textbox)

```python
import tkinter as tk

ventana = tk.Tk()
ventana.title("Ejemplo de Entry")

entrada = tk.Entry(ventana)
entrada.pack()

entrada.get() #Para obtener texto
entrada.insert(0,"Texto inicial") #Para establecer texto inicial

ventana.mainloop()
```

3. üñ±Ô∏è Button ‚Äì Bot√≥n con acci√≥n

```python
import tkinter as tk

ventana = tk.Tk()
ventana.title("Ejemplo de Button")
ventana.geometry("300x200")

def saludar():
    print(f"Hola, Mundo!")

boton = tk.Button(ventana, text="Saludar", command=saludar)
boton.pack()

ventana.mainloop()
```

4. üìã Listbox ‚Äì Lista de elementos


```python
import tkinter as tk

ventana = tk.Tk()
ventana.title("Ejemplo de Listbox")
ventana.geometry("300x200")

lista = tk.Listbox(ventana)
lista.insert(1, "Opci√≥n A")
lista.insert(2, "Opci√≥n B")
lista.pack()

ventana.mainloop()
```

5. üîò Radiobutton ‚Äì Botones de opci√≥n √∫nica

```python
import tkinter as tk

ventana = tk.Tk()
ventana.title("Ejemplo de Radiobutton")
ventana.geometry("300x200")

opcion = tk.StringVar()

tk.Radiobutton(ventana, text="Turno A", variable=opcion, value="A").pack()
tk.Radiobutton(ventana, text="Turno B", variable=opcion, value="B").pack()

ventana.mainloop()
```

6. ‚úÖ Checkbutton ‚Äì Casilla de verificaci√≥n


```python
import tkinter as tk

ventana = tk.Tk()
ventana.title("Ejemplo de Checkbutton")
ventana.geometry("300x200")

estado = tk.BooleanVar()
check = tk.Checkbutton(ventana, text="Activar modo seguro", variable=estado)
check.pack()

ventana.mainloop()
```

7. üì¶ Frame ‚Äì Contenedor para organizar widgets

```python
import tkinter as tk

ventana = tk.Tk()
ventana.title("Ejemplo de Frame")
ventana.geometry("300x200")

marco = tk.Frame(ventana)
marco.pack()

tk.Label(marco, text="Dentro del marco").pack()
tk.Label(ventana, text="Fuera del marco").pack()

ventana.mainloop()
```

8. üñºÔ∏è Imagen - Uso de imagenes en tkinter

```python
import os
import tkinter as tk #Para trabajar con imagenes png, gif
from PIL import Image, ImageTk #Para trabajar con imagenes jpg, jpeg, bmp

#Solo .png o .gif
ruta = os.path.join(os.getcwd(), "img", "images.png")
ruta2 = os.path.join(os.getcwd(), "img", "logo.jpg")
root = tk.Tk()
root.title("Imagenes con Tkinter")
root.geometry("600x480")

# Trabajando con imagenes .png o .gif
imagen = tk.PhotoImage(file=ruta)
label = tk.Label(root, image=imagen)
label.pack()

# Trabajando con imagenes .jpg, .jpeg, .bmp
imagenjpg = Image.open(ruta2)
imagen_tk = ImageTk.PhotoImage(imagenjpg)
label = tk.Label(root, image=imagen_tk)
label.pack()

root.mainloop()
```

üß™ Ejemplo completo: Formulario simple

```python
import tkinter as tk

def mostrar_datos():
    nombre = entrada_nombre.get()
    edad = entrada_edad.get()
    resultado.config(text=f"Nombre: {nombre}, Edad: {edad}")

ventana = tk.Tk()
ventana.title("Formulario")
ventana.geometry("300x200")

tk.Label(ventana, text="Nombre:").pack()
entrada_nombre = tk.Entry(ventana)
entrada_nombre.pack()

tk.Label(ventana, text="Edad:").pack()
entrada_edad = tk.Entry(ventana)
entrada_edad.pack()

tk.Button(ventana, text="Enviar", command=mostrar_datos).pack()
resultado = tk.Label(ventana, text="")
resultado.pack()

ventana.mainloop()
```
