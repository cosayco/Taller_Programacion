# Principales Widgets en Tkinter
***

1. üè∑Ô∏è**Label** ‚Äì Mostrar texto

```python
import tkinter as tk

ventana = tk.Tk()
ventana.title("Ejemplo de Label")

etiqueta = tk.Label(ventana, text="Bienvenido al sistema")
etiqueta.pack()

ventana.mainloop()

2. üìù **Entry** ‚Äì Caja de texto (textbox)


```python
entrada = tk.Entry(ventana)
entrada.pack()

entrada.get() #Para obtener texto
entrada.insert(0,"Texto inicial") #Para establecer texto inicial

3. üñ±Ô∏è Button ‚Äì Bot√≥n con acci√≥n

```python
def saludar():
    nombre = entrada.get()
    etiqueta.config(text=f"Hola, {nombre}!")

boton = tk.Button(ventana, text="Saludar", command=saludar)
boton.pack()

4. üìã Listbox ‚Äì Lista de elementos
```python
lista = tk.Listbox(ventana)
lista.insert(1, "Opci√≥n A")
lista.insert(2, "Opci√≥n B")
lista.pack()


5. üîò Radiobutton ‚Äì Botones de opci√≥n √∫nica
```python
opcion = tk.StringVar()

tk.Radiobutton(ventana, text="Turno A", variable=opcion, value="A").pack()
tk.Radiobutton(ventana, text="Turno B", variable=opcion, value="B").pack()


6. ‚úÖ Checkbutton ‚Äì Casilla de verificaci√≥n
```python
estado = tk.BooleanVar()
check = tk.Checkbutton(ventana, text="Activar modo seguro", variable=estado)
check.pack()


7. üì¶ Frame ‚Äì Contenedor para organizar widgets
```python
marco = tk.Frame(ventana)
marco.pack()

tk.Label(marco, text="Dentro del marco").pack()

üß™ Ejemplo completo: Formulario simple
```python
import tkinter as tk

def mostrar_datos():
    nombre = entrada_nombre.get()
    edad = entrada_edad.get()
    resultado.config(text=f"Nombre: {nombre}, Edad: {edad}")

ventana = tk.Tk()
ventana.title("Formulario")

tk.Label(ventana, text="Nombre:").pack()
entrada_nombre = tk.Entry(ventana)
entrada_nombre.pack()

tk.Label(ventana, text="Edad:").pack()
entrada_edad = tk.Entry(ventana)
entrada_edad.pack()

tk.Button(ventana, text="Enviar", command=mostrar_datos).pack()
resultado = tk.Label(ventana, text="")
resultado.pack()

ventana.mainloop()
